## Simulate diffusion

```python
random_walk=(((np.random.rand(10000)>0.5)-0.5)*2).astype(int)
time=len(random_walk)
```

```python
sns.histplot(random_walk)
```

```python
plt.plot(np.arange(time),np.cumsum(random_walk))
plt.xlabel('Time')
plt.ylabel('Position')
```

```python
abs(np.sum(random_walk))/np.sqrt(time)
```
